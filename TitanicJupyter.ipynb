<h4>Sprawdzono dla kt√≥rych atrybut√≥w wystƒôpujƒÖ brakujƒÖce warto≈õci, w celu odsiania atrbut√≥w zapewniajƒÖcych zbyt ma≈ÇƒÖ ilo≈õƒá informacji.<br>
Atrybuty, dla kt√≥rych brakuje wiƒôcej ni≈º 60 procent pozycji zostanƒÖ usuniƒôte. Zauwa≈ºalnie, w kolumnie z warto≈õciami arybutu "Cabin" istnieje a≈º 76,79 procent brakjƒÖcych warto≈õci. Atrybut zostanie usuniƒôty. <br>
Dla atrybut√≥w "Age" oraz "Embarked" r√≥wnie≈º istnieje pewien procent brakujƒÖcych warto≈õci. W dalszej czƒô≈õci przygotowania zbioru danych zostanƒÖ one naprawione.</h4>

<h4>Zmiana typu danych<br>
Atrybut "Age" widnieje w zadanym zbiorze jako atrybut typu Factor. W ramach czyszczenia danych zosta≈Ç przekonwertowany na atrybut numeryczny</h4>

```r
messyData$Age <- as.numeric(messyData$Age)
```

<h4>Duplikaty<br>

<h4>Po sprawdzeniu, zduplikowane wiersze zosta≈Çy usuniƒôte.</h4>

```r
messyData<-messyData[!duplicated(messyData),]
messyData
```

```r
sum(duplicated(messyData))
```

<h4> Naprawa brakujƒÖcych warto≈õci w atrybutach "Age" oraz "Embarked".<br>
Zauwa≈ºyƒá mo≈ºna, jak do tej pory ustalono, ≈ºe dla atrybutu "Age" istnieje prawie 20 procent brakuƒÖcych warto≈õci w ramach zbioru, a dla atrybutu "Embarked" nie jest to nawet p√≥≈Ç punkta procentowego. Nale≈ºy do obu przypadk√≥w podej≈õƒá indywidualnie.<br>
Braki w atrybucie "Embarked" zostanƒÖ naprawione poprzez uzupe≈Çnienie warto≈õciƒÖ (kategorialnƒÖ) najczƒô≈õciej wystƒôpujƒÖcƒÖ w atrybucie - jest to symbol "S", znacznie przewy≈ºszajƒÖcy liczbowo wystƒÖpienia pozosta≈Çych warto≈õci.</h4>

```r
summary(messyData$Embarked)
```

```r
messyData$Embarked [messyData$Embarked == ''] <- names(which.max(table(messyData$Embarked)))
```

<h4>Dla numerycznego atrybutu "Age" sprawdzono warto≈õƒá najczƒô≈õciej wystƒôpujƒÖcƒÖ oraz medianƒô.</h4>

```r
mean(messyData$Age)
median(messyData$Age)
```

<h4>Wyszukane warto≈õci sƒÖ podobne, braki w atrybucie "Age" zostanƒÖ zamienione na warto≈õƒá najczƒô≈õciej wystƒôpujƒÖcƒÖ w zbiorze.</h4>

```r
messyData$Age [messyData$Age == ''] <- median(messyData$Age)
```

<h4> Po uzupe≈Çnieniu warto≈õci w brakujƒÖcych atrybutach sprawdzono aktualny stan ewentualnych brak√≥w w zbiorze.</h4>

```r
for (i in colnames(messyData)){
  a <- round(sum(messyData[i]=="")/nrow(messyData)*100,2)
  print(i)
  print (a)}
```

<h4>Poprawa liter√≥wek<br>
Zauwa≈ºyƒá mo≈ºna, ≈ºe w≈õr√≥d warto≈õci atrybutu "Sex" wystƒôpujƒÖ b≈Çƒôdne ≈Ça≈Ñcuchy znak√≥w. Docelowo, powinien byƒá to atrybut dychtomiczny - z warto≈õciami "female" oraz "male". </h4>

```r
summary(messyData$Sex)
```

```r
messyData$Sex [messyData$Sex == 'fem'] <- 'female'
messyData$Sex [messyData$Sex == 'femmale'] <- 'female'

messyData$Sex [messyData$Sex == 'mal'] <- 'male'
messyData$Sex [messyData$Sex == 'malef'] <- 'male'
```

```r
summary(messyData$Sex)
```

<h4>Zauwa≈ºono i poprawiono analogicznie wystƒôpujƒÖce liter√≥wki w≈õr√≥d warto≈õci atrybutu "Embarked".<br>
Mo≈ºemy za≈Ço≈ºyƒá, ≈ºe warto≈õƒá "Qe" to liter√≥wka poczyniona podczas wprowadzania symbolu "Q".</h4>

```r
summary(messyData$Embarked)
```

```r
messyData$Embarked [messyData$Embarked == 'Co'] <- 'C'
messyData$Embarked [messyData$Embarked == 'Qe'] <- 'Q'
messyData$Embarked [messyData$Embarked == 'So'] <- 'S'
```

```r
summary(messyData$Embarked)
```

<h4>B≈Çƒôdne znaki<br>
Mo≈ºemy zauwa≈ºyƒá, ≈ºe w≈õr√≥d niekt√≥rch atrybut√≥w, prawdopodobnie przypadkowo, znalaz≈Çy siƒô b≈Çƒôdne warto≈õci. Przyk≈Çadem, w≈õr√≥d warto≈õci atrybutu "Age" mo≈ºemy znale≈∫ƒá warto≈õci ujemne. W ramach analizy za≈Ço≈ºono, ≈ºe symbole "-" oraz "." zosta≈Çy wprowadzone jak liter√≥wki, wiƒôc w ramach czyszczenia zbioru danych podane symbole zostanƒÖ usuniƒôte.</h4>

```r
messyData$Age <- abs(messyData$Age)
```
